```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
selecting_features=['genres','title','cast','director']
```

```python
print(selecting_features)
```

```python
for feature in selecting_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
combine_feature=movies_data['genres']+' '+movies_data['title']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combine_feature)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combine_feature)
```

```python
print(feature_vector)
```

```python
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name=input('Enter your favourite movie name : ')
```

```python
list_titles=movies_data['title'].tolist()
```

```python
print(list_titles)
```

```python
find_close_title=difflib.get_close_matches(movie_name,list_titles)
```

```python
print(find_close_title)
```

```python
close_match=find_close_title[0]
```

```python
index_of_movie=movies_data[movies_data.title == close_match]['index'].values[0]
```

```python
similarity_score=list(enumerate(similarity[index_of_movie]))
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<20):
    print(i,'.',title_from_index)
    i+=1
```
